```python
%pylab inline
from tensorflow import keras
from tensorflow.keras import layers,models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
pwd
```

```python
train_dir=  '/home/elvi/Documents/NNetworks2020/chest_xray' #'C:/ChestXRay2017/chest_xray/train'
```

```python
import os
os.listdir(train_dir)
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,validation_split=0.3)
```

```python
train_generator = train_datagen.flow_from_directory(

    train_dir,
    
    target_size=(150, 150),
    
    batch_size=16,
    
    class_mode='binary',
    
    subset='training')
    
     
dev_generator = train_datagen.flow_from_directory(

            train_dir,
    
            target_size=(150, 150),
    
            batch_size=16,
    
            class_mode='binary',
    
            subset='validation')
    
```

```python
i = 0 
for data_batch, labels_batch in train_generator:
    print('data batch shape:',data_batch.shape)
    print('labels batch shape:',labels_batch.shape)
    i = i + 1
    if i> 5:
        break
```

# 1

Changing the dimensions to 4*4

```python
model = models.Sequential()
model.add(layers.Conv2D(32,(4,4), activation ='relu',input_shape = (150,150,3)))
model.add(layers.MaxPool2D(2,2))
model.add(layers.Conv2D(32,(4,4), activation ='relu'))
model.add(layers.MaxPool2D(2,2))
model.add(layers.Conv2D(32,(4,4), activation ='relu'))
model.add(layers.MaxPool2D(2,2))
#you can add more layers eg.
#model.add(layers.Conv2D(32,(3,3), activation ='relu'))
#model.add(layers.MaxPool2D(2,2))


model.add(layers.Conv2D(32,(4,4), activation ='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation= 'relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile( loss= 'binary_crossentropy',
                 optimizer = optimizers.RMSprop(lr=1e-4), # 0.001, 0.005
                 metrics = ['acc'])
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch = 10,
    verbose = 0,
    epochs = 100, # change epochs: originally 10, 100, 200, 300
    validation_data = dev_generator,
    validation_steps = 10)

#save model:
model.save('case2_v1.0.h5') # every time 0.1, 0.2....
```

```python
ls
```

```python
model.load_weights('case2_v1.0.h5')

#model.load('case2_v1.0.h5')
```

```python
model.summary()
```

```python
accu = history.history['acc'] #training accuracy
val_accu = history.history['val_acc'] #validation accuracy
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accu))
```

```python
plot(epochs, accu, 'bo-', label = 'Training acc')
plot(epochs, val_accu, 'r*-', label = 'Validation acc')
legend()
grid()
show()
```

```python
plot(epochs, loss, 'bo-', label = 'Training loss')
plot(epochs, val_loss, 'r*-', label = 'Validation loss')
legend()
grid()
show()
```
